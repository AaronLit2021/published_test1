```python
import numpy as np
import pandas as pd
import copy
import statistics
```

```python
jobs_df = pd.read_csv("jobs_table.csv")
jobs_list_large = jobs_df.to_dict("records")
jobs = jobs_list_large.copy()
```

```python
def seq_earliest_due_date(jobs_list):
    sequence = sorted(jobs_list, key=lambda d: d['due_date']) 
    return(sequence)
```

```python
def process_sequence(sequence):
    schedule = copy.deepcopy(sequence) 
    t = 0
    for job in schedule:
        proc = job["processing_time"]
        job["start"] = t
        job["complete"] = t + proc
        job["lateness"] = max(0, job["complete"] - job["due_date"])
        if job["lateness"] > 0:
            job["late"] = 1
        else:
            job["late"] = 0
        t += proc
    return(schedule)
```

```python
def evaluate(schedule):
    metrics = {}
    metrics["max_lateness"] = max(job["lateness"] for job in schedule)
    metrics["sum_completion_times"] = sum(job["complete"] for job in schedule) 
    metrics["average_lateness"] = statistics.mean(job["lateness"] for job in schedule)
    metrics["num_jobs_late"] = sum(job["late"] for job in schedule)
    return(metrics)
```
